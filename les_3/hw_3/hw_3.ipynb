Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)

```python
a = np.array([10, 10, 10])
b = np.array([0, 0, -10])
```

```python
def vector_length(vector):
    return np.sqrt(np.sum(np.power(vector, 2)))
```

### 2. Задание
Почему прямые не кажутся перпендикулярными? (см.ролик)


```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

В аналитическом выражении прямые, заданные линейными функциями y = kx + b и y = ax + m будут перпендикулярны, если выполнено следующее условие на их угловые коэффициенты: a*k = -1

```python
print(3 * (-1/3))
```

Условие на угловые коэффициенты выполнено, следовательно заданные в условии прямые перпендикулярны.

Прямые не кажутся перпендикулярными, так как на графике у осей x и y разные пределы. Установим на обоих осях одинаковые пределы значений (-15, 15) и посмотрим на результат.

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlim([-15, 15])
plt.ylim([-15, 15])
plt.xlabel('x')
plt.ylabel('y')
```

После установления на осях одинаковых пределов x и y, прямые на графике выглядят перепендикулярно друг к дургу.

```python
x = []
y1 = []
y2 = []
R = 2
for i in range(1000 * 2 * R + 1):
    x_ = -R + i/1000
    x.append(x_)
    y1.append(math.sqrt(R ** 2 - x_ ** 2))
    y2.append(-math.sqrt(R ** 2 - x_ ** 2))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3, 3)
plt.xlim(-4.3, 4.3)
plt.grid(True)
```

```python
t = np.linspace(0, np.pi, 1000)
x = np.cos(t)
y = np.sin(t)

rho = np.sqrt(x ** 2 + y ** 2)
phi = np.arctan2(y, x) * 180 / np.pi
plt.polar(phi, rho)
```

```python
def xy2pol(x, y): 
    rho = np.sqrt(x ** 2 + y ** 2)
    phi = np.arctan2(y, x) * 180 / np.pi
    return(rho, phi)
    
t = np.linspace(0, np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
rho, phi = xy2pol(x, y)
plt.polar(phi, rho)
```

```python
from scipy.optimize import fsolve
# Решите систему уравнений:
# y = x**2 – 1
# exp(x) + x∙(1 – y) = 1
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) - x*y +x -1)
```

```python
# построим графики
x = np.linspace(-2,3,201)
plt.plot(x,(np.exp(x) +x - 1)/x) #синий
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.5)
plt.grid(True)
plt.show()
```

```python
# решим с-му численным методом
#1
xx = -1.5
yy = 1.5
print("Решение с точкой приближения: (",xx,";",yy,")")
x1, y1 = fsolve(equations,(xx,yy))
print ("x1=",x1,"; y1=",y1)
#2
xx = 2.5
yy = 5.5
print("Решение с точкой приближения: (",xx,";",yy,")")
x2, y2 = fsolve(equations,(xx,yy))
print ("x2=",x2,"; y2=",y2)

print()
```
