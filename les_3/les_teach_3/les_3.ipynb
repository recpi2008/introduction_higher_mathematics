```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,21)
y = 2*x
y_2 = 2*x+1
plt.plot(x,y)
plt.plot(x,y_2)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax= Axes3D(fig)
X=np.arange(-5,5,0.5)
Y=np.arange(-5,5,0.5)
X,Y = np.meshgrid(X,Y)
Z=2*X+5*Y
ax.plot_wireframe(X,Y,Z)
ax.scatter(0,0,0,'z',50,'red')
show()
```

```python
fig = figure()
ax= Axes3D(fig)
X=np.arange(-5,5,0.5)
Y=np.arange(-5,5,0.5)
Z=2*X+3*Y
ax.plot(X,Y,Z,'red')
ax.plot(X,Y,2*Z)
ax.scatter(0,0,0,'z',50,'red')
show()
```

```python
x=np.linspace(-np.pi,np.pi,201)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.xlabel('x')
plt.xlabel('y')
plt.grid(True)
plt.show()
```
