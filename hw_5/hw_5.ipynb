#### 2.2 Сгенерируйте десять выборок случайных чисел х0, …, х9.
и постройте гистограмму распределения случайной суммы  +х0+ …+ х 9. 


```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = []
for _ in range(10000):
    x.append(random.rand(10))

n_bins = 20

x_sum = []
sum_ = 0

for i in range(len(x)-1):
    for j in range(len(x[i])-1):
        sum_ += x[i][j]
    x_sum.append(sum_)
    sum_ = 0

plt.hist(x_sum, n_bins)
plt.show()
```

#### 3.1 Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) и сравните результаты.


```python
import itertools
from math import factorial
```

```python
k , n = 0 , 10000000
a = random.randint(0, 2, n)
b = random.randint(0, 2, n)
c = random.randint(0, 2, n)
d= random.randint(0, 2, n)
x = a + b + c + d
for i in range(0, n):
    if x[i] == 2:
        k += 1
print(k, n, k/n)
```

```python
n , k = 4 , 2
C_n_k = factorial(n) / (factorial(k) * factorial(n - k))
p = 0.5
P_k = C_n_k * p ** k * (1 - p)**(n - k)
print(P_k)
```

#### 3.2 Повторите расчеты биномиальных коэффициентов и вероятностей k успехов в последовательности из n независимых испытаний, взяв другие значения n и k.

```python
n , k = 5 , 1
C_n_k = factorial(n) / (factorial(k) * factorial(n - k))
p = 0.015
P_k = C_n_k * p ** k * (1 - p)**(n - k)
print(P_k)
```

#### 4 Из урока по комбинаторике повторите расчеты, сгенерировав возможные варианты перестановок для других значений n и k


```python
n , k = 10 , 4

print(len(list(itertools.permutations('0123456789', k))))

i = n - 1
A_n_k = n

while i >= (n - k + 1):
    A_n_k *= i
    i -= 1

print(A_n_k)
```

#### 5 Дополните код расчетом коэффициента корреляции x и y по формуле


```python
n = 100
r = 0.7
x = random.rand(n)
y = r*x + (1 - r) * random.rand(n)
plt.plot(x, y, 'o')
plt.grid(True)
plt.show()
print(f'Коэффициент корреляции = {np.corrcoef(x, y)[1,0]}')
```

```python
R = sum((x - np.mean(x)) * (y - np.mean(y))) / np.sqrt(sum((x - np.mean(x))**2) * sum((y - np.mean(y))**2))
print(f'Коэффициент корреляции = {R}')
```
